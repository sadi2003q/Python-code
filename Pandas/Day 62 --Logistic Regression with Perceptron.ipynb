```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
x, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=25)
```

```python
plt.scatter(x[:,0], x[:,1], s=40, c=y, cmap='winter')
```

# Perceptron Trick
A very Basic Version of Finding Logistic regression Value

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perception(x, y):
    x = np.insert(x, 0, 1, axis=1)
    weigh = np.ones(x.shape[1])
    learning_rate = 0.1
    
    for i in range(1000):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(x[j], weigh))
        weigh = weigh+learning_rate * (y[j]-y_hat)*x[j]
    
    return weigh[0], weigh[1:]
```

```python
intercept, coef = perception(x, y)
```

```python
print(intercept)
print(coef)
```

```python
# main equation :  0 = w_1 x_1 + w_2 x_2 + b 
# where:
# 
# 	•	 w_1  and  w_2  are the coefficients (weights) for the features  x_1  and  x_2 
# 	•	 b  is the intercept
# 
# To convert this into the slope-intercept form ( y = mx + b ):

# 1.	Rearrange the equation to solve for  x_2 :
    # x_2 = -\frac{w_1}{w_2} x_1 - \frac{b}{w_2} 
# 2.	Compare this with the standard line equation  y = mx + b :
    # •	The slope ( m ) is given by  -\frac{w_1}{w_2} 
    # •	The intercept ( b ) is given by  -\frac{b}{w_2} 

m = -(coef[0] / coef[1])
b = -(intercept / coef[1])

x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
```

```python

plt.plot(x_input, y_input, color='red', linewidth=4)
plt.scatter(x[:,0], x[:,1], s=40, c=y, cmap='winter')
plt.ylim(-3,2)
```

# Applying Logistic Regression From Scikit Learn

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(x, y)
```

```python
m = -(logistic_regression.coef_[0][0]/ logistic_regression.coef_[0][1])
b = -(logistic_regression.intercept_/logistic_regression.coef_[0][1])
```

```python
x_input2 = np.linspace(-3, 3, 100)
y_input2 = m*x_input2 + b
```

```python
plt.plot(x_input, y_input, color='red', linewidth=4)
plt.plot(x_input2, y_input2, color='black', linewidth=4)
plt.scatter(x[:,0], x[:,1], s=40, c=y, cmap='winter')
plt.ylim(-3,2)
```
