```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight_height.csv')
```

```python
df.shape
```

```python
figure, (fig2, fig3) = plt.subplots(ncols = 2, figsize = (12, 5))

df['Height'].plot(kind='box', ax = fig2)
df['Weight'].plot(kind='box', ax = fig3)
```

# Percentile Limit difining

```python
upper_limit = df['Height'].quantile(0.99)
lower_limit = df['Height'].quantile(0.01)
```

# Trimming

```python
Trimming_df = df[(df['Height']>=lower_limit) & (df['Height']<=upper_limit)]
```

```python
Trimming_df['Height'].plot(kind='box')
```

# Capping

```python
capping_df = df
```

```python
capping_df['Height'] = np.where(
    capping_df['Height']<lower_limit,
    lower_limit,
    np.where(
        capping_df['Height']>upper_limit,
        upper_limit,
        capping_df['Height']
    )
)
```
