```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

from sklearn.model_selection import train_test_split as split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score as score

from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import cross_val_score as score
```

```python
df = pd.read_csv('concrete_data.csv')
```

```python
df.info()
```

```python
x = df.drop(columns = ['Age'])
y = df['Age']
```

```python
train_x, test_x, train_y, test_y = split(x, y, test_size= 0.2)
```

# Without using Preprocessing

```python
lr = LogisticRegression()
```

```python
lr.fit(train_x, train_y)
```

```python
pred_y = lr.predict(test_x)
```

```python
print('Accuracy Score : ', score(pred_y, test_y))
```

```python
for col in df :
  figure, (fig1, fig2) = plt.subplots(ncols=2, figsize=(5, 2))
  sns.distplot(df[col], ax = fig1, hist = False)
  stats.probplot(df[col], plot=fig2)
  plt.show()
```

# With Using Preprocessing

```python
power = PowerTransformer()
```

```python
new_train_x = power.fit_transform(train_x)
new_test_x = power.transform(test_x)
```

```python
transform_train_x = pd.DataFrame(new_train_x)
transform_test_x = pd.DataFrame(new_test_x)
```

```python
new_lr = LogisticRegression()
```

```python
new_lr.fit(transform_train_x, train_y)
```

```python
pred_y = new_lr.predict(transform_test_x)
```

```python
print('Accuracy Score : ', score(pred_y, test_y))
```

```python
for col in transform_train_x :
  figure, (fig1, fig2) = plt.subplots(ncols=2, figsize=(5, 2))
  sns.distplot(transform_train_x[col], ax = fig1, hist = False)
  stats.probplot(transform_train_x[col], plot=fig2)
  plt.show()
```
