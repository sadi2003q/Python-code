```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
x, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=25)
```

```python
plt.scatter(x[:,0], x[:,1], s=40, c=y, cmap='winter')
```

# Logistic Regression

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(x, y)

intercept = logistic_regression.intercept_
coefficients = logistic_regression.coef_

m = -(coefficients[0][0]/coefficients[0][1])
b = -(intercept/coefficients[0][1])

x_input = np.linspace(3, -3, 100)
y_input = m*x_input + b
```

```python
plt.plot(x_input, y_input, color = 'yellow', linewidth = 4)
plt.scatter(x[:,0], x[:,1], s=40, c=y, cmap='winter')
plt.ylim(-3,2)
```

# User Define Logistic Regression

```python
def my_regression(x, y):
    x = np.insert(x, 0, 1, axis = 1)
    weight = np.ones(x.shape[1])
    learning_rate = 0.5
    
    for i in range(5000):
        y_hat = sigmoid(np.dot(x, weight))
        weight = weight + learning_rate * (np.dot((y-y_hat), x))/x.shape[1]
        
    return weight[0], weight[1:]
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
intercept, coefficients = my_regression(x, y)
```
