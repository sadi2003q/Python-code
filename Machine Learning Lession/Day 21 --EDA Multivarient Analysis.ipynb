```python
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
bill = pd.read_csv('Bill.csv')
fuel = pd.read_csv('Fuel.csv')
```

# Scatterplot (Numerical - Numerical)

```python
bill.head()
```

```python
sns.scatterplot(x=bill['ave_monthly_income'], y=bill['amount_paid'], hue=bill['is_ac'], style=bill['num_children'], size= bill['num_people'])
```

```python
sns.barplot(x=fuel['FUEL'], y=fuel['FUEL CONSUMPTION'], hue=fuel['VEHICLE CLASS'])
```

# Box plotting

```python
sns.boxplot(x=train['Survived'], y= train['Age'], hue=train['Sex'])
```

# Dist plotting 

```python
sns.distplot(train[train['Survived']==1]['Age'],hist = False)
sns.distplot(train[train['Survived']==0]['Age'], hist = False)
```

```python
sns.distplot(train[train['Sex']=='male']['Survived'],  hist = False)
sns.distplot(train[train['Sex']=='female']['Survived'],  hist = False)
```

# heat Map

```python
heat_map = pd.crosstab(train['Sex'], train['Survived'])
sns.heatmap(heat_map)
```

```python
heat_map
```

```python
engine_consumption = pd.crosstab(fuel['VEHICLE CLASS'], fuel['FUEL CONSUMPTION'])
engine_consumption
```

# Cluster Map

```python
sns.clustermap(pd.crosstab(train['Pclass'], train['Survived']))
```

# Pair Plot

```python
sns.pairplot(train, hue='Fare')
```

# Line Plot

```python
sns.lineplot(x=bill['num_people'], y=bill['ave_monthly_income'])
```
