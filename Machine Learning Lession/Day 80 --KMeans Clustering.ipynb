# Loading Dataset

```python
from sklearn.datasets import load_digits 
```

```python
digit = load_digits()
```

```python
x = digit.data
y = digit.target
```

# Model Making 

```python
from sklearn.cluster import KMeans
```

```python
cluster = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    cluster.append(km.inertia_)
```

# Model Plotting

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1, 11), cluster)
```

# Final Algorithm

```python
model = KMeans(n_clusters=9)
y_pred = model.fit_predict(x)
```

```python
from sklearn.decomposition import PCA

# Reduce the data to 2 dimensions for visualization
pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x)

# Plot the clusters
plt.scatter(x_reduced[:, 0], x_reduced[:, 1], c=y_pred, cmap='viridis', marker='.')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('KMeans Clustering (9 clusters)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Reduce the data to 3 dimensions for visualization
pca = PCA(n_components=3)
x_reduced = pca.fit_transform(x)

# Plot the clusters in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x_reduced[:, 0], x_reduced[:, 1], x_reduced[:, 2], c=y_pred, cmap='viridis', marker='o')

ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
ax.set_title('KMeans Clustering (9 clusters)')

# Add a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Cluster')

plt.show()
```

```python
import plotly.express as px
# Reduce the data to 3 dimensions for visualization
pca = PCA(n_components=3)
x_reduced = pca.fit_transform(x)

# Create a DataFrame for plotly
import pandas as pd

df = pd.DataFrame(x_reduced, columns=['PCA Component 1', 'PCA Component 2', 'PCA Component 3'])
df['Cluster'] = y_pred

# Plot the clusters in 3D using plotly
fig = px.scatter_3d(df, x='PCA Component 1', y='PCA Component 2', z='PCA Component 3',
                    color='Cluster', title='KMeans Clustering (9 clusters)',
                    color_continuous_scale='Viridis')

fig.show()
```
