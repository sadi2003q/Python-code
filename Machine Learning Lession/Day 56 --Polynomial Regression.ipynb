```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split as split

from sklearn.preprocessing import PolynomialFeatures 

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
plt.plot(x, y, 'b. ')
```

```python
train_x, test_x, train_y, test_y = split(x, y, test_size = 0.2, random_state = 2)
```

```python
model = LinearRegression()
```

```python
model.fit(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
r2_score(y_pred, test_y)
```

```python
plt.plot(train_x,model.predict(train_x),color='r')
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
poly = PolynomialFeatures(degree=4, include_bias=True)
```

```python
new_train_x = poly.fit_transform(train_x)
new_test_x = poly.transform(test_x)
```

```python
new_model = LinearRegression()
```

```python
new_model.fit(new_train_x, train_y)
```

```python
y_pred = new_model.predict(new_test_x)
```

```python
r2_score(y_pred, test_y)
```

```python
plt.plot(train_x, new_model.predict(new_train_x), 'g. ', color='r')
plt.plot(x, y, "b.")
```

```python
sgd = SGDRegressor()
```

```python
sgd.fit(new_train_x, train_y)
```

```python
y_pred = sgd.predict(new_test_x)
```

```python
r2_score(y_pred, test_y)
```

```python
# 
# plt.plot(x, y, "b.")
plt.plot(train_x, sgd.predict(new_train_x), 'b. ', color='r')
```
