# Random Forest
The Random Forest algorithm is an ensemble learning method primarily used for classification and regression tasks. It operates by constructing multiple decision trees during training and outputting the mode of the classes (classification) or mean prediction (regression) of the individual trees. Here are the key characteristics of Random Forest

# Loading Dataset

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('All_CSV/Student_performance_data _.csv')
```

```python
df = df.drop(columns=['StudentID'])
```

# heat Map

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr(), annot=True)
```

# using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from Model_Training import Model_Training
```

```python
model_random_forest = Model_Training(RandomForestClassifier(n_estimators=500, max_depth=5, max_samples=0.5, max_features=0.5, random_state=42), df.drop(columns=['GradeClass']), df['GradeClass'])
```

```python
score=model_random_forest.run()
```

```python
score.get_score()
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_decision_tree = Model_Training(DecisionTreeClassifier(max_depth=5), df.drop(columns=['GradeClass']), df['GradeClass'])
```

```python
score_decision_tree = model_decision_tree.run()
```

```python
score_decision_tree.get_score()
```
